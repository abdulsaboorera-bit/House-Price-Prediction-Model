```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('/content/homepricesm.csv')
df
```

```python
df.bedrooms.median()
```

```python
df.bedrooms = df.bedrooms.fillna(df.bedrooms.median())
df
```

```python
df.bedrooms
```

```python
plt.scatter(df.area,df.price,color="green")
```

```python
x=df.drop("price",axis=1)
y=df.price
```

```python
x
```

```python
y
```

```python
model=linear_model.LinearRegression()
model.fit(x,y)
```

```python
model.score(x,y)
```

```python
model.predict([[3300,3,5]])
```

```python
model.coef_
```

#Age is in negative beacuse it is inverse to area
